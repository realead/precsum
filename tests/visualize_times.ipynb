```python
import numpy as np
import perfplot
import precsum as ps

def ps_pairwise(x):
    res = np.empty(x.shape[1], dtype=np.float32)
    ps.pairwise_sum_2d(x,res,axis=0)
    
def ps_pairwise_v2(x):
    res = np.empty(x.shape[1], dtype=np.float32)
    ps.pairwise_sum_2d_v2(x,res,axis=0)
        
def np_sum(x):
    x.sum(axis=0)
    
```

```python
perfplot.show(
    setup=lambda n: np.ones((n,100), dtype=np.float32),
    n_range=[2**k for k in range(0,20)],
    kernels=[
        np_sum, 
        ps_pairwise_v2,
        ps_pairwise, 
        ],
    logx=True,
    logy=True,
    xlabel='#rows',
    title="100 cols non-contiguous memory, axis=0",
    equality_check = None
    )
```
